### <font color='green'>Desafío 1: Generación de funciones</font>
* Genere funciones para calcular la media y varianza de un vector. Debe cumplir con los siguientes requisitos:
 * Ambas funciones deben ingresar un argumento **x** correspondiente al vector.
 * Las funciones deben contener `docstrings` con la documentación asociada a la variable y retornar un valor utilizando `return`.
 * La función de la varianza se debe llamar a la función de la media.
* Utilice las funciones para reportar la información sobre `goles_favor`, `goles_contra` y `puntos`.

```python
def _mean(vector):
    """
        Objetivo:
            - Retornar la media de un vector entregado
            
        Parámetros:
            - vector (Series): Vector con los datos (lista, Series). Obligatorio.

        Retorno:
           - media (float): La media de los datos del vector entregago
    """
    #Se calcula la media del vector
    media = vector.sum()/len(vector)
    
    return media
    
    
def _var(vector):
    """
        Objetivo:
            - Retornar la varianza de un vector entregado
            
        Parámetros:
            - vector (DataFrame): Vector con los datos (lista, Series). Obligatorio.

        Retorno:
           - var (float): La varianza de los datos del vector entregado
    """
    media = _mean(vector)
    suma = [(i-media)**2 for i in vector]
    var = sum(suma)/(len(vector)-1)
    return(var)
```

```python
print(f'''Media:
La columna goles_favor tiene una media igual a {_mean(df['goles_favor']):.2f}
La columna goles_contra tiene una media igual a {_mean(df['goles_contra']):.2f}
La columna puntos tiene una media igual a {_mean(df['puntos']):.2f}

Varianza:
La columna goles_favor tiene una varianza igual a {_var(df['goles_favor']):.2f}
La columna goles_contra tiene una varianza igual a {_var(df['goles_contra']):.2f}
La columna puntos tiene una varianza igual a {_var(df['puntos']):.2f}
''') 
```

### <font color='green'>Desafío 2:</font>

Utilizando el método `groupby` de la clase `DataFrame`, en conjunto con la función `.agg` de
pandas, calcule la media, la varianza y desviación estándar de la cantidad de goles a favor,
en contra y de la cantidad de puntos por continente.
 * ¿En qué continente se observa una mayor cantidad de goles a favor?
 * ¿En qué continente se observa una mayor cantidad de goles en contra?
 * ¿En qué continente se observa una mayor cantidad de puntos en promedio?


```python
#Creamos la agrupacion agregada con los campos solicitados y lo mostramos en pantalla
df_group = df.groupby(['continent'])[['goles_favor','goles_contra','puntos']].agg(['mean','var','std'])
display(df_group)

#Se obtiene el nombre del continente donde se observa las mayores cantidades solicitadas
max_gf = df_group[df_group['goles_favor']['mean'] == df_group['goles_favor']['mean'].max()].index[0]
max_gc = df_group[df_group['goles_contra']['mean'] == df_group['goles_contra']['mean'].max()].index[0]
max_puntos = df_group[df_group['puntos']['mean'] == df_group['puntos']['mean'].max()].index[0]

print(f'''
En el continente {max_gf} se observa la mayor cantidad de goles a favor
En el continente {max_gc} se observa la mayor cantidad de goles en contra
En el continente {max_puntos} se observa la mayor cantidad de puntos promedio
''') 
```

### <font color='green'>Desafío 3: Simulaciones</font>
* Genere una función `generate_pet` que devuelva de forma aleatoria un string `'perro'`o `'gato'` un número n de veces. Ejecútela un par de veces.
* Aplique la función `generate_pet` para generar 20 muestras.
* ¿Cuál es la probabilidad de elegir un perro al azar? ¿Y un gato?
* Agregue `np.random.seed(2)` al inicio del chunk. ¿Qué diferencia hay cuando se ejecuta la función varias veces luego de fijar la semilla?

```python
def generate_pet(n_muestra):
    """
        Objetivo:
            - Retornar un dataframe con una columna que tenga como valor perro o gato
            
        Parámetros:
            - n_muestra (int): numero de elementos total del dataframe. Obligatorio.

        Retorno:
           - valor (list): retorna el dataframe con valores perro o gato
    """
    
    string = ['perro','gato']
    lista = [np.random.choice(string, p=[0.5, 0.5]) for i in range(n_muestra)]
    df = pd.DataFrame(lista, columns = ['animal'])
    
    return df
```

```python
lista = generate_pet(20).value_counts(normalize = True)
lista
```

```python
print(f'''
La probabilidad de elegir un perro al azar es de {lista.get("perro")*100:.0f}%
La probabilidad de elegir un gato al azar es de {lista.get("gato")*100:.0f}%
''') 
```

Finalmente, al agregar el `np.random.seed(2)` dentro de la función, lo que hace es que parta con la misma semilla con el fin de que al momento de aleatorizar, siempre salga el misma resultado de la primera ejecutacion. Esto porque parte con la misma semilla cuando por dentro genera el numero aleatoria para el calculo de las robabilidades

### <font color='green'>Desafío 4: Función simuladora</font>
* Genere una función llamada `simulate_pets_prob` que tome como argumento un número finito de simulaciones a generar.
* La función debe simular dos situaciones `young_pet` y `old_pet`, y contar la ocurrencia de los siguientes casos:
 * De las dos mascotas simuladas (young y old), contar las ocasiones donde por lo menos una de las mascotas sea un perro.
 * De las dos mascotas simuladas, contar las ocasiones donde `old_pet` sea un perro.
 * De las dos mascotas simuladas, contar las ocasiones donde los dos sean perros.
* El método debe tener una semilla pseudoaleatoria de 1.
* El output de la función debe entregarse en términos de Probabilidad.
* De los tres escenarios, ¿Cuál es el menos probable? ¿Cuál es el más probable? ¿Por qué?

```python
def simulate_pets_prob(numero_simulaciones):
    """
        Objetivo:
            - Retornar la media de un vector entregado
            
        Parámetros:
            - vector (Series): Vector con los datos (lista, Series). Obligatorio.

        Retorno:
           - media (float): La media de los datos del vector entregago
    """
    np.random.seed(1)
    young_pet = generate_pet(numero_simulaciones).rename(columns={'animal':'young_pet'})
    old_pet = generate_pet(numero_simulaciones).rename(columns={'animal':'old_pet'})
    df = pd.concat([young_pet, old_pet], axis =1)
    
    #primer caso
    count1 = df[(df['young_pet']=='perro') | (df['old_pet']=='perro')].shape[0]
    prob1 = count1/len(df)*100
    
    #Segundo caso
    count2 = df[df['old_pet']=='perro'].shape[0]
    prob2 = count2/len(df)*100
    
    #Tercer caso
    count3 = df[(df['young_pet']=='perro') & (df['old_pet']=='perro')].shape[0]
    prob3 = count3/len(df)*100
    
    asdf = df.value_counts()
    
    return df, prob1, prob2, prob3, asdf
```

```python
df, prob1, prob2, prob3, asdf = simulate_pets_prob(20)
display(df)

print(f'''
De los 3 escenarios:
* ¿Cuál es el menos probable?: El escenario menos probable es donde en las dos columnas deban haber un perro, 
con una probabilidad de {prob3}%
* ¿Cuál es el más probable?: El escenario más probable es donde en al menos en alguna columna haya un perro, 
con una probabilidad de {prob1}%
* ¿Por qué?: El tercer caso siempre será el que tenga la menor probabilidad, esto debido a que ese espacio muestral
está contenido en los otros dos casos, por ende, su probabilidad siempre será menor o igual a los demas. Por otro 
lado, el primer caso siempre será el caso con mayor probabilidad, esto debido a que los otros dos espacios muestrales 
estan contenido dentro del primer caso, por ende, la probabilidad del primer caso, siempre será mayor o igual a los 
demás casos.
'''
)
```
